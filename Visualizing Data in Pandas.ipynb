# Panda Visualization

```python
# Import essential libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical computing and operations
import matplotlib.pyplot as plt  # For creating visualizations and plots
```

```python
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(r"D:\Data Analyst Course\Python Projects\Panda Visualization\Ice Cream Ratings.csv")

df  # Display the DataFrame
```

```python
# Print all available Matplotlib styles
print(plt.style.available)

# Apply the 'tableau-colorblind10' style for better color differentiation
plt.style.use('tableau-colorblind10')
```

```python
# Create a line plot to visualize Ice Cream Ratings over time
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# Create a stacked bar plot to visualize the distribution of values across different categories
df.plot(kind = 'bar', stacked = True)
```

```python
# Create a stacked bar plot for the "Flavor Rating" column
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# Create a stacked horizontal bar plot to visualize the distribution of values across categories
df.plot.barh(stacked = True)
```

```python
# Create a scatter plot to visualize the relationship between "Texture Rating" and "Overall Rating"
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Red')
```

```python
# Create a histogram to visualize the distribution of numerical data
df.plot.hist(bins = 20)
```

```python
# Create a box plot to visualize the distribution of numerical columns in the DataFrame
df.boxplot()
```

```python
# Create an area plot to visualize trends over time or categories
df.plot.area(figsize = (10,5))
```

```python
# Create a pie chart to visualize the distribution of "Flavor Rating"
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
